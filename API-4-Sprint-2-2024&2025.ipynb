```python
arquivo_1 = '/content/drive/My Drive/DadosAntaq/2024_2025/2024Atracacao.csv'
arquivo_2 = '/content/drive/My Drive/DadosAntaq/2024_2025/2024Carga.csv'
arquivo_3 = '/content/drive/My Drive/DadosAntaq/2024_2025/2024TemposAtracacaoParalisacao.csv'
arquivo_4 = '/content/drive/My Drive/DadosAntaq/2024_2025/2025Atracacao.csv'
arquivo_5 = '/content/drive/My Drive/DadosAntaq/2024_2025/2025Carga.csv'
arquivo_6 = '/content/drive/My Drive/DadosAntaq/2024_2025/2025TemposAtracacaoParalisacao.csv'
```

```python
Atracacao24 = pd.read_csv(arquivo_1, sep=';', encoding='utf-8', low_memory=False)
Carga24 = pd.read_csv(arquivo_2, sep=';', encoding='utf-8', low_memory=False)
TemposParalisacao24 = pd.read_csv(arquivo_3, sep=';', encoding='utf-8', low_memory=False)
Atracacao25 = pd.read_csv(arquivo_4, sep=';', encoding='utf-8', low_memory=False)
Carga25 = pd.read_csv(arquivo_5, sep=';', encoding='utf-8', low_memory=False)
TemposParalisacao25 = pd.read_csv(arquivo_6, sep=';', encoding='utf-8', low_memory=False)
```

```python
Atracacao24_25 = pd.concat([Atracacao24, Atracacao25], ignore_index=True)
Carga24_25 = pd.concat([Carga24, Carga25], ignore_index=True)
TemposParalisacao24_25 = pd.concat([TemposParalisacao24, TemposParalisacao25], ignore_index=True)
```

```python
Atracacao24_25.info()
```

```python
Atracacao24_25.head(10)
```

```python
cols_datas = [
    "Data Atracação", "Data Chegada", "Data Desatracação",
    "Data Início Operação", "Data Término Operação"
]

for col in cols_datas:
    Atracacao24_25[col] = pd.to_datetime(Atracacao24_25[col], errors="coerce", dayfirst=True)
```

```python
cols_texto = [
    "Porto Atracação", "Complexo Portuário", "Terminal",
    "Município", "UF", "Região Geográfica"
]

for col in cols_texto:
    Atracacao24_25[col] = Atracacao24_25[col].str.strip().str.upper()
```

```python
Carga24_25 = pd.concat([Carga24, Carga25], ignore_index=True)
```

```python
Carga24_25.info()
```

```python
Carga24_25.head(10)
```

```python
cols_texto = [
    "Origem", "Destino", "Tipo Operação da Carga",
    "Carga Geral Acondicionamento", "Tipo Navegação",
    "Natureza da Carga", "Sentido"
]

for col in cols_texto:
    Carga24_25[col] = Carga24_25[col].str.strip().str.upper()
```

```python
Carga24_25["VLPesoCargaBruta"] = pd.to_numeric(
    Carga24_25["VLPesoCargaBruta"].str.replace(",", ".", regex=False),
    errors="coerce"
)
```

```python
Carga24_25["VLPesoCargaBruta"].isna().sum()
```

```python
Carga24_25["TEU"] = pd.to_numeric(
    Carga24_25["TEU"].str.replace(",", ".", regex=False),
    errors="coerce"
)

cols_flags = [
    "FlagCabotagem", "FlagCabotagemMovimentacao",
    "FlagLongoCurso", "FlagOffshore"
]

for col in cols_flags:
    Carga24_25[col] = Carga24_25[col].astype("Int64")
```

```python
Carga24_25["IDCarga"].duplicated().sum()
```

```python
TemposParalisacao24_25 = pd.concat([TemposParalisacao24, TemposParalisacao25], ignore_index=True)
```

```python
TemposParalisacao24_25.info()
```

```python
TemposParalisacao24_25.tail(10)
```

```python
for col in ["DTInicio", "DTTermino"]:
    TemposParalisacao24_25[col] = pd.to_datetime(
        TemposParalisacao24_25[col],
        errors="coerce"
    )
```

```python
TemposParalisacao24_25["Duracao"] = (
    TemposParalisacao24_25["DTTermino"] - TemposParalisacao24_25["DTInicio"]
)
TemposParalisacao24_25["Duracao_horas"] = (
    TemposParalisacao24_25["Duracao"].dt.total_seconds() / 3600
)
```

```python
TemposParalisacao24_25["DescricaoTempoDesconto"] = (
    TemposParalisacao24_25["DescricaoTempoDesconto"]
    .str.strip()
    .str.upper()
)
```

```python
TemposParalisacao24_25[TemposParalisacao24_25["Duracao_horas"] < 0]
```

```python
TemposParalisacao24_25["Duracao_horas"] = TemposParalisacao24_25["Duracao_horas"].round(2)
```

```python
Atracacao24_25.to_csv("Atracacao2024_2025.csv", index=False, sep=";", encoding="utf-8")
Carga24_25.to_csv("Carga2024_2025.csv", index=False, sep=";", encoding="utf-8")
TemposParalisacao24_25.to_csv("TemposParalisacao2024_2025.csv", index=False, sep=";", encoding="utf-8")
```
