```python
arquivo1 = '/content/drive/My Drive/DadosAntaq/ParAtracCar_2024.csv'
arquivo2 = '/content/drive/My Drive/DadosAntaq/ParAtracCar_2025.csv'
arquivo3 = '/content/drive/My Drive/DadosAntaq/Instalacao_Origem.csv'
arquivo4 = '/content/drive/My Drive/DadosAntaq/Instalacao_Destino.csv'
```

```python
PAC2024 = pd.read_csv(arquivo1, sep=';', encoding='utf-8', low_memory=False)
PAC2025 = pd.read_csv(arquivo2, sep=';', encoding='utf-8', low_memory=False)
Insta_Orig = pd.read_csv(arquivo3, sep=';', encoding='utf-8', low_memory=False)
Insta_Dest = pd.read_csv(arquivo4, sep=';', encoding='utf-8', low_memory=False)
```

```python
Insta_Orig.info()
```

```python
Insta_Orig.head(10)
```

```python
colunas_necessarias = [
    'Origem',
    'Origem Nome',
    'UF.Origem',
    'País Origem',
    'Continente Origem'
]

Insta_Orig = Insta_Orig[colunas_necessarias].copy()
```

```python
for col in colunas_necessarias:
    Insta_Orig[col] = (
        Insta_Orig[col]
        .astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )
Insta_Orig['Origem'] = Insta_Orig['Origem'].astype(str)
```

```python
Insta_Orig = Insta_Orig.drop_duplicates(subset='Origem')
```

```python
Insta_Dest = pd.read_csv(arquivo4, sep=';', encoding='utf-8', low_memory=False)
```

```python
Insta_Dest.info()
```

```python
Insta_Dest.head(10)
```

```python
colunas_necessarias = [
    'Destino',
    'Nome Destino',
    'UF.Destino',
    'País Destino',
    'Continente Destino'
]

Insta_Dest = Insta_Dest[colunas_necessarias].copy()
```

```python
for col in colunas_necessarias:
    Insta_Dest[col] = (
        Insta_Dest[col]
        .astype(str)
        .str.lower()
        .str.strip()
        .str.replace(r'\s+', ' ', regex=True)
    )
Insta_Dest['Destino'] = Insta_Dest['Destino'].astype(str)
```

```python
Insta_Dest = Insta_Dest.drop_duplicates(subset='Destino')
```

```python
PAC2024 = pd.read_csv(arquivo1, sep=';', encoding='utf-8', low_memory=False)
PAC2025 = pd.read_csv(arquivo2, sep=';', encoding='utf-8', low_memory=False)
```

```python
PAC2024_2025 = pd.concat([PAC2024, PAC2025], ignore_index=True)
```

```python
PAC2024_2025.info()
```

```python
PAC2024_2025["Origem"] = PAC2024_2025["Origem"].str.upper()
PAC2024_2025["Destino"] = PAC2024_2025["Destino"].str.upper()
```

```python
Insta_Orig["Origem"] = Insta_Orig["Origem"].str.upper()
Insta_Dest["Destino"] = Insta_Dest["Destino"].str.upper()
```

```python
PAC2024_2025["Origem"] = PAC2024_2025["Origem"].str.strip()
PAC2024_2025["Destino"] = PAC2024_2025["Destino"].str.strip()
```

```python
Insta_Orig = Insta_Orig.drop_duplicates("Origem")
Insta_Dest = Insta_Dest.drop_duplicates("Destino")
```

```python
PAC2024_2025 = PAC2024_2025.merge(Insta_Orig, on="Origem", how="left")
```

```python
PAC2024_2025 = PAC2024_2025.merge(Insta_Dest, on="Destino", how="left")
```
